```python
def percent_consumption(country,year1, year2):
    import matplotlib.pyplot as plt
    import csv
    growth = []
    years = [i for i in range(1961, 2023)]
    reader = csv.reader(open('API_NE.CON.TOTL.ZS_DS2_en_csv_v2_6229714.csv','r'), delimiter = ',')
    for i in reader:
        if country in i:
            growth = growth = [float(j) for j in i[4:-2]]
    
    new = zip(years, growth)
    new_growth =[]
    new_years = list(range(year1,year2 +1))
    for i in new:
        if i[0] >= year1 and i[0] <= year2:
            new_growth.append(i[1])
            
    fig, ax = plt.subplots(figsize=(11,8))
    ax.plot(new_years, new_growth)
    plt.ylabel("GDP % Consumption")
    plt.xlabel('year')
    plt.title(f'Consumption as of GDP % in period {year1} to {year2} for {country}')
```

```python
percent_consumption("China",2010, 2022)
```
